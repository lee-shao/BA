```python
# Funktion zum Erstellen der Boxplots
# def plot_boxplots(df_results, metric, clf, output_dir):
def plot_boxplots(df_results, metric, output_dir):
    plt.figure(figsize=(14, 8))
    # filtered_df = df_results[df_results['classifier'] == clf]
    # sns.boxplot(x='k', y=metric, data=filtered_df)
    sns.boxplot(x='k', y=metric, data=df_results)
    plt.title(f'Boxplot of {metric} by k values')
    plt.xlabel('k')
    plt.ylabel(metric)
    plt.xticks(rotation=90, fontsize=8)
    plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.7)
    plt.savefig(os.path.join(output_dir, f'boxplot_{metric}.png'))
    plt.show()
    plt.close()
```

```python
classifiers = ['lr', 'rf', 'gb']
metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc']
output_dir = 'C:/Users/leesh/Desktop/BA/Notebooks/results_(a)_BU_1000/results_(a)_BU_w_alco_1000_(16.08)_x10/boxplots'
os.makedirs(output_dir, exist_ok=True)

for metric in metrics:
    plot_boxplots(df_results, metric, output_dir)

# for clf in classifiers:
    # for metric in metrics:
        # plot_boxplots(df_results, metric, clf, output_dir)
```
